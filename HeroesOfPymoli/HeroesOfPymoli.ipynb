```python
import pandas as pd
import os
import numpy as np
```

```python
filepath=os.path.join("purchase_data.json")
purchase_data_df = pd.read_json(filepath)
purchase_data_df.head()
# purchase_data_df.count() #Just checked that there weren't null values
```

```python
#Player Count
player_count = purchase_data_df["SN"].nunique()
columns = {'Total Player Count': player_count}
player_count_df = pd.DataFrame([columns])
player_count_df.head()
```

```python
#Sort by unique screen names and genders#Sort b 
age_data = purchase_data_df.loc[:, ("Age", "SN")]
age_data_df = age_data.sort_values("Age")

#Create Bins
age_bins = [0, 10, 14, 18, 22, 26, 30, 34, 38, 42, 99]
age_labels = ['06-10', '11-14', '15-18', '19-22', '23-26', '27-30', '31-34', '35-38', '39-42', '42+']
age_groups = pd.cut(age_data_df["Age"], age_bins, labels=age_labels)

#add age groups to dataframe
age_data_df["Age Group"] = age_groups

#organize dataframe by age groups
age_data_df2 = age_data_df.groupby(["Age Group"])["SN"].nunique()
age_data_df2 = pd.DataFrame(age_data_df2)
age_data_df2 = age_data_df2.rename(columns={"SN":"Number of Players"})
# age_data_df2

#Calculate percents and add to dataframe
age_group_counts = age_data_df2.loc[:, "Number of Players"]
age_percent = age_group_counts/player_count*100
age_cols = {'Count': age_group_counts, 'Percent': age_percent}
age_demographics_df = pd.DataFrame(age_cols)
age_demographics_df.head()
```

```python
# **Top Spenders**

# * Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#   * SN
#   * Purchase Count
#   * Average Purchase Price
#   * Total Purchase Value
#Count number of purchases by each screen name
sn_count = purchase_data_df["SN"].value_counts()

#Group data by screen name and sum of purchases by that person
sn_tot_purchases = purchase_data_df.groupby("SN")["Price"].sum()

#Make it a data frame
sn_tot_purchases2 = pd.DataFrame(sn_tot_purchases)

#Add count and create average
sn_tot_purchases2["Number of Purchases"] = sn_count
price_sum = sn_tot_purchases2.loc[:, "Price"]
sn_tot_purchases_df = sn_tot_purchases2.sort_values("Price", ascending=False)

sn_tot_purchases_df["Average Purchase Price"] = price_sum / sn_count
sn_tot_purchases_df = sn_tot_purchases_df.rename(columns={"Price":"Total Purchases"})

sn_tot_purchases_df["Average Purchase Price"] = sn_tot_purchases_df["Average Purchase Price"].map("${:.2f}".format)
sn_tot_purchases_df["Total Purchases"] = sn_tot_purchases_df["Total Purchases"].map("${:.2f}".format)

sn_tot_purchases_df.head()
```

```python
# **Most Popular Items**

# * Identify the 5 most popular items by purchase count, then list (in a table):
#   * Item ID
#   * Item Name
#   * Purchase Count
#   * Item Price
#   * Total Purchase Value

#count by item name #count  
most_popular_items = purchase_data_df["Item ID"].value_counts()

#create a data frame
pop_items = purchase_data_df.loc[:, ('Item ID', 'Price', 'Item Name')]

#sort by item id and number purchased
pop_items2 = pop_items.groupby(["Item ID"])["Item Name"].value_counts()
pop_items2 = pd.DataFrame(pop_items2)

#sort by item id and total value purchased
pop_items3 = pop_items.groupby(["Item ID"])["Price"].sum()
pop_items3 = pd.DataFrame(pop_items3)

#join lists
pop_items4 = pop_items2.join(pop_items3, how='outer')

#rename columns
pop_items4 = pop_items4.rename(columns={"Item Name":"Number Purchased", "Price":"Total Purchase Value"})

#find individual item price
tot_price = pop_items4.loc[:, "Total Purchase Value"]
number_purchased = pop_items4.loc[:, "Number Purchased"]
item_price = tot_price/number_purchased

pop_items4["Item Price"] = item_price
pop_items_final = pop_items4.sort_values("Number Purchased", ascending=False)

pop_items_final["Item Price"] = pop_items_final["Item Price"].map("${:.2f}".format)
pop_items_final["Total Purchase Value"] = pop_items_final["Total Purchase Value"].map("${:.2f}".format)
pop_items_final.head()
```

```python
#**Most Profitable Items**
most_profitable_items = pop_items_final.sort_values("Total Purchase Value", ascending=False)
most_profitable_items.head()
```
